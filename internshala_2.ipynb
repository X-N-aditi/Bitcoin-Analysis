```python
# plotting bar plot of bitcoin "open", "high", "low" ,"close"

sns.barplot(data=df[["Open", "High", "Low", "Close"]])
plt.title("Bitcoin Price Distribution")
plt.grid(True)
```

```python
# visualizing Moving-Average of bitcoin of 7 days and 30 days 

df["MA_7"]=df["Close"].rolling(window=7).mean()
df["MA_30"]=df["Close"].rolling(window=30).mean()

plt.figure(figsize=(12, 6))
plt.plot(df["Date"], df["Close"], label="Close Price", alpha=0.5)
plt.plot(df["Date"], df["MA_7"], label="7-Day MA", color="red")
plt.plot(df["Date"], df["MA_30"], label="30-Day MA", color="green")

plt.xlabel("Date")
plt.ylabel("Price")
plt.title("Bitcoin Closing Price with Moving Averages")
```

```python
# visualizing Volume of Bitcoin 
plt.figure(figsize=(12, 5))
plt.bar(df['Date'], df['Volume'], color='green')
plt.title("Bitcoin Trading Volume")
plt.xlabel("Date")
plt.ylabel("Volume")
```

```python
df_period = df[(df['Date'] >= '2015-01-01') & (df['Date'] <= '2025-01-20')]
df_period['Year'] = df_period['Date'].dt.year
yearly_growth = df_period.groupby('Year')['Close'].last().pct_change() * 100

print(yearly_growth)
```

* so bitcoin rate is increasing day by day
* bitcoin rate is highest on January month
* bitcoin rate after increasing day year after year got falls on 2019 (reason : pendemic)
* after pendemic the highest rise on 20 january 2025
* so investing money on bitcoin is beneficial becuase year by year its increasing

```python
import numpy as np
from sklearn.linear_model import LinearRegression
```

## make some predictions

```python
df_period['Timestamp'] = df_period['Date'].astype('int64') // 10**9  # Convert Date to UNIX timestamp

X = df_period['Timestamp'].values.reshape(-1, 1) 
y = df_period['Close']

# Train the model
model = LinearRegression()
model.fit(X, y)

future_dates = pd.date_range(start='2026-01-01', periods=5, freq='Y')  
future_timestamps = future_dates.astype('int64') // 10**9  

predictions = model.predict(future_timestamps.values.reshape(-1, 1))

for date, price in zip(future_dates, predictions):
    print(f"{date.date()} : {price:.2f}")
```

<h1 style='color: red;'> Bitcoin prediction 2026-2030</h1>

```python
plt.figure(figsize = (12, 5))

plt.plot(df_period['Date'], df_period['Close'], color='blue', label='Actual Price', markersize=1)
plt.plot(future_dates, predictions, color='red', linestyle= '-', marker='o', label='Predicted price')
plt.xlabel('Year')
plt.ylabel('Bitcoin price')
plt.title('Bitcoin Price prediction 2026-2030')
plt.grid()
plt.legend()
```
